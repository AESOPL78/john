```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Arial Unicode MS'
```

```python
import glob 
csv = glob.glob('./서울특별시 공공자전거 대여이력 정보_22.06*.csv')
csv
```

```python
rent_2022=pd.DataFrame()   #2022 데이터 만들기
for file in csv:
        temp = pd.read_csv(file, encoding='cp949')
        rent_2022=pd.concat([rent_2022, temp])
rent_2022.head()
```

```python
csv=glob.glob('서울특별시 공공자전거 대여이력 정보_2306*.csv')
csv
```

```python
rent_2023=pd.read_csv(csv[0],encoding='cp949')  #2023 데이터만들기
rent_2023.head()
```

```python
bike_rent=pd.concat([rent_2022,rent_2023])  #2022데이터와 2023 데이터 합침
bike_rent.head(1599999)
```

```python
file=glob.glob('공공자전거 대여소 정보_수정(23.06월 기준).*xlsx')
file
```

```python
bike_station = pd.read_excel(file[0])
bike_station
```

```python
bike_station.isna().sum()
```

```python
col_name= ['자치구','대여소ID','상세주소','위도','경도','거치대수']
```

```python
rent_name=['대여 대여소_구','대여 대여소번호', '대여 대여소주소','대여 위도','대여 경도', '대여 거치대수']
rent_station= bike_station[col_name]
rent_station.columns= rent_name
```

```python
return_name= ['반납 대여소_구','반납대여소번호', ' 반납대여소주소','반납 위도', '반납 경도','반납 거치대수']
return_station=bike_station[col_name]
return_station.columns= return_name
```

```python
bike_rent=pd.merge(bike_rent,rent_station, on='대여 대여소번호')
bike_rent.head()
```

```python
bike_rent=pd.merge(bike_rent,return_station,on='반납대여소번호')
bike_rent.head(0)
```

```python
bike_rent['대여일시']
```

```python
bike_rent['대여일시']=pd.to_datetime(bike_rent['대여일시'])
bike_rent['반납일시']=pd.to_datetime(bike_rent['반납일시'])
```

```python
bike_rent['대여 년도']=bike_rent['대여일시'].dt.year
bike_rent['대여 일']=bike_rent['대여일시'].dt.day
bike_rent['대여 요일']=bike_rent['대여일시'].dt.day_name()
bike_rent['대여 시간']=bike_rent['대여일시'].dt.hour
bike_rent.head(1)
```

```python
bike_rent['반납 년도']=bike_rent['반납일시'].dt.year
bike_rent['반납 일']=bike_rent['반납일시'].dt.day
bike_rent['반납 요일']=bike_rent['반납일시'].dt.day_name()
bike_rent['반납 시간']=bike_rent['반납일시'].dt.hour
bike_rent.head(1)
```

```python
station_rent= bike_rent.groupby(['대여 대여소명'])[['대여일시']].count()
station_rent.reset_index(inplace=True)         #인덱스 초기화 
station_rent
```

```python
target_col= ['대여 위도','대여 경도']
station_loc= bike_rent.groupby('대여 대여소명')[target_col].first()  #시각화를 위해 대여소를 위도와 경도 표시함 
station_loc.reset_index(inplace=True)
station_loc
```

```python
marker= pd.merge(station_rent, station_loc, on='대여 대여소명')
marker
```
