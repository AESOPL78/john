```python
df_list=[] #반복문을 통해 전체파일을 로드함
for file_path in file_names:
  df_temp=pd.read_csv(file_path, low_memory=False)
  df_list.append(df_temp)
```

```python
#하나의 데이터프레임으로 합치기 #(행개수,  컬럼개수)
df=pd.concat(df_list)
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
#사용할 칼럼만 남기기
df=df[['상호명','상권업종대분류명','시도명','시군구명','도로명주소','경도','위도']].copy()
df.shape
```

```python
df.info()
```

```python
#상호명 없는데이터 제거
df=df[df["상호명"].notnull()].copy()
df.shape
```

```python
#상호명의 영문을 대문자로 변경 - 상호명 편하게 가져오기위함
df["상호명_대문자"]=df["상호명"].str.upper()
```

```python
#버거킹,맥도날드,KFC,롯데리아 자료 추출
burger=["버거킹","맥도날드","KFC","롯데리아"]
```

```python
df=df[df["상호명_대문자"].str.contains("버거킹|Burger|맥도날드|멕도날드|롯데리아|KFC|케이에프씨")]
```

```python
df.shape
```

```python
#상호명 전처리를 위해 유일값 확인
df["상호명"].unique()[:50]
```

```python
df.loc[df["상호명_대문자"].str.contains("롯데리아"),"브랜드"]="롯데리아"
df
```

```python
#브랜드명 파생변수 만들기
for bname in burger :
  df.loc[df["상호명_대문자"].str.contains(bname), "브랜드"]= bname
```

```python
df.loc[df["상호명_대문자"].str.contains("케이에프씨"),"브랜드"]="KFC" #상호명 오류방지
```

```python
df.loc[df["상호명_대문자"].str.contains("멕도날드"),"브랜드"]="맥도날드"
```

```python
df[["상호명","브랜드"]] #확인용
```

```python
df["브랜드"].value_counts().plot.bar(rot=0)  #데이터속 롯맥버케 갯수 비교
```

```python
#전국 시도별 상호수와 버거지수
df_skorea = df.groupby(["시도명","브랜드"])["상호명"].count().unstack().fillna(0).astype(int)
df_skorea["합계"]= df_skorea.sum(axis=1)
df_skorea
```

```python
#버거지수계산
df_skorea["버거지수"]=(
    df_skorea["KFC"] + df_skorea["버거킹"]+ df_skorea["맥도날드"])/df_skorea["롯데리아"]
df_skorea
```

```python
df_city_latlog = df.groupby(["시도명"]).agg({"위도":"mean","경도":"mean"}) #위도 경도 구하기
df_city_latlog
```

```python
df_skorea_latlong= df_city_latlog.join(df_skorea) #위경도+버거지수
df_skorea_latlong
```

```python
df_dist_count = df.groupby(
    ["시도명","시군구명","브랜드"])["상호명"].count().unstack().fillna(0).astype(int)
df_dist_count["합계"]=df_dist_count.sum(axis=1)
df_dist_count
```

```python
df_dist_count["버거지수"]=(df_dist_count["KFC"]+df_dist_count["버거킹"]+df_dist_count["맥도날드"])/ df_dist_count["롯데리아"]
df_dist_count
```

```python
#시군구별 버거지수
df_dist_count["버거지수"]=(df_dist_count["KFC"]+ df_dist_count["버거킹"]+df_dist_count["맥도날드"])/df_dist_count["롯데리아"]
df_dist_count
```

```python
import folium
```

```python
latlong = df.loc[df["시도명"]=="서울특별시", ["위도","경도"]].median().values
latlong
```

```python
df_seoul=df[df["시도명"]=="서울특별시"]
```

```python
folium.Map(location=[37.54571109, 127.00484684])
```

```python
f_map=folium.Map(latlong, zoom_start=12, tiles="OpenStreetMap")
for i in df_seoul.index:
    sub_lat=df_seoul.loc[i, "위도"]
    sub_long=df_seoul.loc[i, "경도"]

    brand= df_seoul.loc[i,'브랜드']
    title= f"{df_seoul.loc[i,'상호명']}-{df_seoul.loc[i,'도로명주소']}"
    color={"롯데리아": "yellow", "버거킹":"blue","맥도날드":"orange","KFC":"red"}

    folium.CircleMarker([sub_lat,sub_long],radius=5,color=color[brand],tooltip=title).add_to(f_map)
f_map
```
